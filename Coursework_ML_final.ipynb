# 2. Исследование зависимостей и формулирование гипотез

```python
Итак, предварительно почищенные данные готовы. Гипотезы мы будем формулировать относительно целевого признака "Churn". Он является категориальным, значит для выявления взаимосвязей между ним и остальными признаками используем коэффициент корреляции Кендалла. Но предварительно надо ранжировать все категориальные признаки. Для этого пизнаки, имеющие две категории, закодируем с помощью метода доступа cat.codes, а признаки с числом категорий более двух - с помощью one-hot encoding.
```

```python
# Кодируем бинарные признаки
obj_cols = telecom_df.select_dtypes(['object']).columns

for col in obj_cols:
    if len(list(telecom_df[col].unique())) <= 2:
        print(f'В столбце {col}:  ', end = '')
        cat_col = telecom_df[col].astype('category')
        codes = cat_col.cat.codes
        cats = cat_col.cat.categories
        telecom_df[col] = codes
        for i, cat in enumerate(cats):
            if i == len(cats) - 1:
                print(f'{cat} --> {i}.')
            else:
                print(f'{cat} --> {i},  ', end = '')
```

```python
# Кодируем признаки с числом категорий более двух
telecom_df = pd.get_dummies(telecom_df)
telecom_df
```

Очевидно, что столбцы, относящиеся к No_internet_service будут абсолютно одинаковы. Оставим только один из них и переименуем его в No_internet

```python
# Удалим дублирующиеся признаки
telecom_df = telecom_df.drop(['OnlineSecurity_No internet service', 'OnlineBackup_No internet service', \
                             'DeviceProtection_No internet service', 'TechSupport_No internet service', \
                             'StreamingTV_No internet service'], axis = 1)
telecom_df = telecom_df.rename (columns= {'StreamingMovies_No internet service': 'No_internet'})
telecom_df
```

Построим гистограммы для непрерывных переменных tenure и MonthlyCharges для случаев, когда клиент остался и когда ушел.

```python
# Гистограммы непрерывных переменных
fig, axs = plt.subplots(figsize = (20, 8), nrows = 1, ncols = 2)
for ax, col in zip(axs.flatten(), ['tenure', 'MonthlyCharges']):
    snsplot = sns.kdeplot(telecom_df[telecom_df['Churn'] == 1][col], shade=True, label = 'Churn = 1', ax = ax)
    snsplot = sns.kdeplot(telecom_df[telecom_df['Churn'] == 0][col], shade=True, label = 'Churn = 0', ax = ax)
    ax.set_title(f'Распределенние признака {col}', y = 1.05, color = 'green', size = 20)
    ax.set_xlabel(f'{col}', color = 'blue', size = 16)
    ax.set_ylabel('Частота', color = 'blue', size = 16)
    ax.legend()
fig.tight_layout();
```

Как мы видим, с ростом ежемесячных платежей количество ушедших увеличивается. Этот признак сильно влияет на целевую переменную. Но самое интересное - это распределение оттока в зависимости от срока пользования услугами оператора. И мы видим крайне интересную картину - пик оттока приходится на крайне малые сроки пользования оператором. Вспомним то, что мы выяснили в пункте 1 нашего исследования - четверть клиентов пользуются услугами оператора менее 10 месяцев. А теперь мы видим, что отток среди них максимальный. Именно эту группу нам надо удержать в первую очередь. Давайте построим график зависимости количества ушедших клиентов в зависимости от срока использования услуг. Разобъем признак tenure на интервалы и в каждом интервале посчитаем количество ушедших. Построим графики для разных значений количества интервалов.

```python
# Поиск точки перехода графика из крутой части в пологую
fig, axs = plt.subplots(figsize = (16, 16), nrows = 3, ncols = 3)
for i, ax in enumerate(axs.flatten()):
    tenure_df = telecom_df[telecom_df['Churn'] == 1].copy()
    tenure_df['new'] = pd.cut(tenure_df['tenure'], bins=10 + i * 10)
    tenure_df = pd.DataFrame(tenure_df.groupby('new')['Churn'].count()).reset_index()
    tenure_df['new'] = tenure_df['new'].apply(lambda x: x.mid)
    ax.set_title(f'Зависимость оттока клиентов\nот времени пользования услугами\
    компании\nпри количестве интервалов bins = {10 + i * 10}', y = 1.1, color = 'green')
    ax.plot(tenure_df['new'].values, tenure_df['Churn'].values)
    ax.set_xlabel('Время использования услуг оператора, месяцев', color = 'blue')
    if i in [0, 3, 6]:
        ax.set_ylabel('Количество ушедших, человек', color = 'blue')
fig.tight_layout()
plt.show()
```

Разное количество интервалов мы использовали, чтобы поточнее определить визуально с помощью графиков точку перехода от экстремального оттока клиентов к более ровной части графиков. Также мы замечаем еще одну интересную вещь - при приближении срока пользования услугами оператора к максимальному, зафиксированному в данном датасете, отток тоже начинает резко увеличиваться.

Итак, используя графики, мы видим, что экстремальнные оттоки клиентов происходят до срока пользования услугами оператора 10 месяцев. Некоторые варианты дальнейших гипотез построим на разделении всего исходного датафрейма на две части в зависимости от значения фичи "tenure": [0, 10] месяцев и (10, 72] месяцев.

## Гипотеза № 1

Первую гипотезу построим на всем наборе почищенных данных - используем датафрейм **telecom_df**.

Для второй гипотезы уберём самые малосвязанные с целевой переменной признаки. Для этого используем методы Кендалла и Спирмана. В качестве порога для значимости возьмем значение 0.1. Все признаки с силой связи с целевой переменной ниже этого порога уберем из данных

```python
# Вычисление маловлияющих на целевую переменную признаков
corr_1 = telecom_df.corr(method='kendall')['Churn'].rename('Churn_1')
corr_2 = telecom_df.corr(method='spearman')['Churn'].rename('Churn_2')
corr_df = pd.concat([corr_1, corr_2], axis=1)
features_to_del = corr_df[(abs(corr_df['Churn_1']) < 0.1) & (abs(corr_df['Churn_2']) < 0.1)]
features_to_del
```

```python
# Удаление маловлияющих на целевую переменную признаков
cols_to_del = list(features_to_del.index)
telecom_df_1 = telecom_df.drop(cols_to_del, axis = 1)
telecom_df_1
```

Построим матрицу парных корреляций.

```python
# Матрица парных корреляций,построенная на коэффициенте Кендалла
corr_m = telecom_df_1.corr(method='kendall')

colormap = plt.cm.RdBu
plt.figure(figsize=(20,20))
plt.title('Матрица корреляции (коэффициент Кендалла)', y=1.05, size=20, c = 'g')
sns.heatmap(corr_m,linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True);
```

Для анализа корреляции категориальных данных также подходит коэффициент Спирмана. Давайте построим матрицу и для него.

```python
# Матрица парных корреляций,построенная на коэффициенте Спирмана
corr_m = telecom_df_1.corr(method='spearman')

plt.figure(figsize=(20,20))
plt.title('Матрица корреляции (коэффициент Спирмана)', y=1.05, size=20, c = 'g')
sns.heatmap(corr_m,linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True);
```
